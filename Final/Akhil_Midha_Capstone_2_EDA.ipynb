```python
df = pd.read_csv('../Data.csv')
```

```python
df['Year']=df['Year'].astype(str)
df['Year']=[x[:4] for x in df['Year']]
df['Year']=pd.to_datetime(df['Year'],format='%Y')
```

```python
df.head()
```

```python
ax = sns.scatterplot(data=df, x='Year', y='Total').set(title='year vs total CO2 emissions')
```
